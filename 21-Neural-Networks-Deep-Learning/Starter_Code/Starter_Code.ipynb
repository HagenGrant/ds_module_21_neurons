```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
```

```python
# Import and read the charity_data.csv.
df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN'
df = df.drop(columns = ["EIN", "SPECIAL_CONSIDERATIONS"], axis=1)
df.head()
```

```python
# Determine the number of unique values in each column.
df.nunique(axis=0)
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
df.APPLICATION_TYPE.value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
threshold = 500
application_type_counts = df.APPLICATION_TYPE.value_counts()
application_types_to_replace = application_type_counts.loc[application_type_counts < threshold].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    df['APPLICATION_TYPE'] = df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
df.CLASSIFICATION.value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts = df.CLASSIFICATION.value_counts()
classification_counts.loc[classification_counts > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
threshold = 1000
classifications_type_counts = df.CLASSIFICATION.value_counts()
classifications_to_replace = classifications_type_counts.loc[classifications_type_counts < threshold].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    df['CLASSIFICATION'] = df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
df['CLASSIFICATION'].value_counts()
```

```python
##Lets try grouping some of the NAMEs togethers that are greater than 5
name_counts = df['NAME'].value_counts()
name_counts[name_counts>5]
```

```python
# Determine which values to replace if counts are less than or equal to 5
names_to_replace = list(name_counts[name_counts <= 5].index)

# Replace in dataframe
for app in names_to_replace:
    df['NAME'] = df['NAME'].replace(app,"Other")

# Check to make sure replacement was successful
df['NAME'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
cat_features = ["NAME","APPLICATION_TYPE", "AFFILIATION", "CLASSIFICATION", "USE_CASE", "ORGANIZATION", "INCOME_AMT"]
df_cat = df.loc[:, cat_features]
df_cat["NAME"] = df_cat.APPLICATION_TYPE.astype(str)
df_cat["APPLICATION_TYPE"] = df_cat.APPLICATION_TYPE.astype(str)
df_cat["AFFILIATION"] = df_cat.AFFILIATION.astype(str)
df_cat["CLASSIFICATION"] = df_cat.CLASSIFICATION.astype(str)
df_cat["USE_CASE"] = df_cat.USE_CASE.astype(str)
df_cat["ORGANIZATION"] = df_cat.ORGANIZATION.astype(str)
df_cat["INCOME_AMT"] = df_cat.INCOME_AMT.astype(str)
#df_cat["SPECIAL_CONSIDERATIONS"] = df_cat.SPECIAL_CONSIDERATIONS.astype(str)

df_cat = pd.get_dummies(df_cat, dtype=int)
```

```python
##Putting everything back together
not_cat_features = ["STATUS", "ASK_AMT", "IS_SUCCESSFUL"]
df_final = df.loc[:, not_cat_features]
df_final = pd.concat([df_final, df_cat], axis=1)
df_final.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = df_final.drop(["IS_SUCCESSFUL"], axis='columns').values
y = df_final["IS_SUCCESSFUL"].values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
df_final.head()
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=100, activation="relu", input_dim=len(X_train[0])))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=90, activation="sigmoid"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=80, activation="sigmoid"))


# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))



# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=500)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
